```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

```python
#loading the data from file in pandas dataframe
car_dataset=pd.read_csv('../data/car details v4.csv')
```

```python
car_dataset = car_dataset.dropna()  #drop missing values
```

Encoding The Categorical Data

```python
car_dataset = car_dataset.drop(['Model'], axis=1) #Removes Model Column from car_dataset and store the updated dataset back to car_dataset
```

```python
#Select Categorical Columns(Python list created and stores in categorical_cols variable
categorical_cols = ['Make', 'Fuel_Type', 'Transmission', 'Location', 'Color','Owner', 'Seller_Type', 'Engine', 'Max Power', 'Max Torque', 'Drivetrain']
```

```python
# Encode all categorical columns at once
car_dataset_encode=pd.get_dummies(car_dataset,columns=categorical_cols,drop_first=True)
```

```python
print(car_dataset.columns.tolist())
```

Splitting The Data And Target

```python
X=car_dataset.drop(['Price'],axis=1) #Droping Price column and assigning dataset to X variable
Y=car_dataset['Price']  #Selecting only Price column and assigning to Y variable
print((Y))
```

Splitting Training And Test Data

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,random_state=2) #Spliting of dataset into training and testing data where X--> Input Features Y--> Target Variable(What you want to predict
```

```python
#Hot Encoding
X_train = pd.get_dummies(X_train, drop_first=True)
X_test = pd.get_dummies(X_test, drop_first=True)

# Align train and test (important if some categories exist in train but not test)
X_test = X_test.reindex(columns=X_train.columns, fill_value=0)
```

Modal Training


1.Linear Regression

```python
print(X_train.isnull().sum().sum())  #prints sum of  missing values
print(X_test.isnull().sum().sum())   #prints sum of  missing values
```

```python
#loading the linear regression modal
lin_reg_modal=LinearRegression()
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_train.dtypes)  # Ensure all numeric
print(X_train.isna().sum())  # Check for missing values
print(Y_train.isna().sum())   # Check for missing values
```

```python
lin_reg_modal=lin_reg_modal.fit(X_train,Y_train) #Trains Model using training data
```

Modal Evalaution

```python
#Prediction On Training Data
training_data_prediction=lin_reg_modal.predict(X_train)
```

```python
# R square Error
error_score= metrics.r2_score(Y_train,training_data_prediction)
print(error_score)
```

Visualize the actual proces and predicted prices

```python
plt.scatter(Y_train,training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```

```python
#Prediction On Testing Data
test_data_prediction=lin_reg_modal.predict(X_test)
```

```python
# R square Error
error_score= metrics.r2_score(Y_test,test_data_prediction)
print(error_score)
```

```python
plt.scatter(Y_test,test_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```
